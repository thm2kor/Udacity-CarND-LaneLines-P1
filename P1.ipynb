## Ideas for Lane Detection Pipeline

**Some OpenCV functions (beyond those introduced in the lesson) that might be useful for this project are:**

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=3):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """    
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)


def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, extrapolate = False):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    if extrapolate:
        draw_lines_extended (line_img, lines)
    else:
        draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
   
````

## Helper functions

The helper functions are used as-is. Only exception is the `hough_lines` function. To support extrapolation, the function is adapted to include an additional flag. The adapted `draw_lines` function is available later in this space.

The find_lane function configures the parameters required for Canny and Hough transform algorithms. The region of interest is a fixed trapezoid derived by empirically examining the sample images and videos. The lane finding involves six sequential steps (as discussed in the class notes). 
1. The given image is converted to gray scale
2. The image is further smoothened and Canny algorithm is run on it
3. A fixed region of interest mask is applied to the resulting image
4. Hough transform are applied to the edges to identify straight lines
5. The end points of the slopes are collected in a array. A linear regression with these points to identify the coordinates of the lane line. 
6. Draw the identified lane lines on the target image

````python
def find_lane (image, extrapolate = False):
    ## Canny Algorithm parameters are anchored to the values discussed in the Udacity course
    kernel_size = 3         # For Gaussian smoothing internal to Canny algorithm
    low_threshold = 50      # Canny algo rejects pixels below this threshold
    high_threshold = 150    # Canny algo detects strong  pixels above this threshold

    ## Hough transform parameters are adapted using trial and error methods on the 6 given images
    rho =  1                # distance in pixels of the grid in Hough space
    theta = np.pi/90        # angular resolution in radians of the grid in Hough space
    threshold = 20          # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 15    # minimum length in pixels that would qualify as a line
    max_line_gap = 10       # maximum gap in pixels between segments

    # Pull out the x and y sizes and make a copy of the image
    ysize = image.shape[0]
    xsize = image.shape[1]
    color_select = np.copy(image)
  
    ysize = image.shape[0]
    xsize = image.shape[1]

    # Define a trapezoid region of interest
    # I choose a "fixed" trapezoid (based on the images in the pipeline) to capture the region of interest.
    # potential shortcoming of this approach. May not work for other scenarios

    left_bottom = (0,ysize)                    
    right_bottom = (xsize,ysize)               
    right_top =  ((xsize/2)+50, (ysize/2 + 60))
    left_top = ((xsize/2 )-50, (ysize/2 + 60))

    vertices = np.array([[left_bottom,left_top,right_top, right_bottom]], dtype=np.int32)
    
    # convert the given image to gray scale 
    gray = grayscale(color_select)
    
    # apply gaussian smoothing for suppressing noise and spurious gradients by averaging
    blur_gray = gaussian_blur(gray,kernel_size)

    # run canny algorithm
    edges = canny(blur_gray, low_threshold, high_threshold)

    # mask out the pixels in our region of interest
    masked_edges = region_of_interest (edges, vertices)

    # Run Hough on edge detected image
    lines = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap , extrapolate)
    
    # draw the lines from hough transform on the source image
    result = weighted_img(lines, image)

    return result
````

Now for the one with the solid yellow lane on the left. This one's more tricky!

````python
def draw_lines_extended (img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: The end-points (x1,y1,x2,y2) of each line within the lines array are consolidated in individual arrays. 
    A linear fit y = mx+c function is generated based on the X and Y points.
    The resulting slope(m) and y-intercept(c) is used to extrapolate the co-ordinates for each lane.
    """
    imshape = img.shape
    x_left = []
    y_left = []
    x_right = []
    y_right = []
    y_min = img.shape[0]
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            if (x1 == x2 ) or (y1 == y2):
                continue
            slope = (y2-y1)/(x2-x1)
            # filter lines which are horizontal or the slope does not fit the slope of the land
            # This "hard-coding" could be a potential drawback of this approach.
            # Slope derived based on the examination of slopes from the given images. 
            if (slope > -0.5 and slope < 0.5) or slope < -1 or slope > 1:
                continue
            if slope < 0:
                x_left.append(x1)
                y_left.append(y1)
                x_left.append(x2)
                y_left.append(y2)
            elif slope > 0:
                x_right.append(x1)
                y_right.append(y1)
                x_right.append(x2)
                y_right.append(y2)
                
   
    if len(x_left) and len(y_left): 
        # Linear regression to fit the identified points on the left lane
        [m,c] = np.polyfit(x_left, y_left, 1)
        x_min = np.int(np.float((y_min - c)) / np.float(m))
        x_max = np.int(np.float((min(y_left) - c)) / np.float(m))
        cv2.line(img, (x_min, y_min), (x_max, min(y_left)), color, thickness)
    
    if len(x_right) and len(y_right):
        # Linear regression to fit the identified points on the right lane
        [m,c] = np.polyfit(x_right, y_right, 1)
        x_min = np.int(np.float((y_min - c)) / np.float(m))
        x_max = np.int(np.float((min(y_right) - c)) / np.float(m))
        cv2.line(img, (x_min, y_min), (x_max, min(y_right)), color, thickness)
````
